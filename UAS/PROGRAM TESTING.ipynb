```python
import pandas as pd
data_1000 = pd.read_csv('pilkada-1000-uas.csv')
data_100 = pd.read_csv('pilkada-100-uas.csv')

data_1000.head()
```

```python
data_100.head()
```

```python
# Drop Kolom tabel yang tidak digunakan
unused_col = ['Jumlah Paslon', 'Nama Kandidat (1)', 'Nama Kandidat (2)','Usia (1)', 'Usia (2)', 'Petahana? (1)', 'Petahana? (2)']

def drop(data):
    data = data.drop(unused_col, axis=1)
    return data

data_1000 = drop(data_1000)
data_100 = drop(data_100)
```

```python
data_1000.head()
```

```python
# Convert Dataset ke Integer
from sklearn.preprocessing import LabelEncoder

def labeling(data):
  for col in data_1000.columns.values:
    if data[col].dtypes == 'object':
      data[col] = LabelEncoder().fit_transform(data[col])
  return data

data_1000 = labeling(data_1000)
data_100 = labeling(data_100)
```

```python
x_train = data_1000.drop('Hasil Pilkada', axis = 1)
x_test = data_100.drop('Hasil Pilkada', axis = 1)
y_train = data_1000['Hasil Pilkada']
y_test = data_100['Hasil Pilkada']
```

```python
# Extract Decision Tree
from sklearn.tree import DecisionTreeClassifier, export_text

# (max_depth=1000, criterion='gini', ccp_alpha=0.001, random_state=42, min_samples_split=10)
classifier = DecisionTreeClassifier(max_depth=1000, criterion='gini', ccp_alpha=0.001, random_state=42)
classifier.fit(x_train, y_train)
```

```python
# Kalkulasi Skor Predictions
from sklearn.metrics import accuracy_score

predictions = classifier.predict(x_test)
accuracy = accuracy_score(y_test, predictions)

print(f'Tingkat Akurasi: {accuracy:.2f} / {100 * accuracy:.2f} %')
```

```python
# Kalkulasi Matriks Konfusi dan Hasil Klasifikasi
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
print(classification_report(y_test, predictions))
print(confusion_matrix(y_test, predictions))
```

```python
# Decision Tree Visualization
tree_rules = export_text(classifier, feature_names=list(data_1000.columns[1:]))

print(tree_rules)
```

```python
# Decision Tree Image Visualization
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(15, 10))
_ = plot_tree(classifier, filled=True, ax=ax)
plt.show()
```
